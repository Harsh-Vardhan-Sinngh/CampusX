```python
 def foo(gruop):
  return gruop['Series_Title'].str.startswith('A').sum()
```

```python
def rank_movie(group):
  group['genre_rank']=group['IMDB_Rating'].rank(ascending=False)
  return group
```

```python
genre.apply(rank_movie)
```

```python
def normal(group):
  group['norm_rating'] = (group['IMDB_Rating']-group['IMDB_Rating'].min())/(group['IMDB_Rating'].max()-group['IMDB_Rating'].min())
  return group
```

```python
genre.apply(normal)
```

```python
ipl=pd.read_csv('/content/deliveries (1).csv')
ipl.head(2)
```

```python
ipl.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False)
```

```python
six=ipl[ipl['batsman_runs']==6]
six
```

```python
six.groupby('batsman')['batsman'].count()
```

```python
six.groupby('batsman')['batsman'].count().sort_values(ascending=False)
```

```python
temp_df=ipl[ipl['over']>15]
temp_df[(temp_df['batsman_runs']==4) | (temp_df['batsman_runs']==6)]
temp_df.groupby('batsman')['batsman'].count().sort_values(ascending=False)
```

```python
temp=ipl[ipl['batsman']=='V Kohli']
temp.groupby('bowling_team')['batsman_runs'].sum().reset_index()
```

```python
temp=ipl[ipl['batsman']=='V Kohli']
temp.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).head(1).values[0]
```

```python
def highest(batsman):
  temp=ipl[ipl['batsman']==batsman]
  return temp.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).head(1).values[0]
```

```python
highest('DA Warner')
```
